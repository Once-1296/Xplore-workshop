### Example 1: Checking if number is Even/Odd and Positive/Negative/Zero

````python
num = int(input("Enter a number: "))

if num > 0:
    print("The number is positive.")
elif num < 0:
    print("The number is negative.")
else:
    print("The number is zero.")

# 6 lines is too big tho, here's a Pythonic way to do it in just 1 !
print("The number is" + (" positive" if num > 0 else " negative" if num < 0 else " zero"))
# same result, Neat !
````

````python
# switch case of C++ is replaced by match case
option = -1
match option:
  case 1:
    print("You chose option 1!")
  case 2:
    print("You chose option 2!")
  case _: # default choice
    print("You chose neither option!")
````

## **Loops**

### For loops
- As you know, for loops are primarily: start, perform stuff, go to next value and end
- That is the basis of for loops. which gets really simplified in Python
- If You Are dealing with an iterable , you just do `for value in iterable:`
- For example: `for x in arr:`
- While for accessing ranges, we have `for i in range(start,stop,step):`
- `start` is inclusive while `stop` is exclusive ie `[start,stop)`
- Here only `stop` is necessary argument as `start` has default value 0 and `step` has default value 1
- We can also use `reversed(arr)` or `reversed(range(...))` to iterate from end instead of start


````python
# for loops are pretty similar, utilising indentation instead of curly braces !
arr = [1,4,6,7] # random array
sum = 0
for num in arr:
  sum+=num
  print("number:", num)
print("sum: ",sum)
````

````python
# lets double each element in arr, though in a more pythonic for loop
doubled_arr = [2*x for x in arr]
print("doubled array: ",doubled_arr)
````

````python
# what about printing all even numbers below 10?
print("even numbers below 10: ")
for i in range(10):
  if i%2==0: print(i,end=" ")
print("\n")
# here range(10) covers [0,10), ie 0 to 9
````

## **Data Types (Type Checking and Type Casting)**

### Type Checking
Determine the type of a variable using `type()`.

### Type Casting (Type Conversion)
Change a variable from one type to another.

- **Implicit Type Casting:** Python automatically converts data types during operations.
- **Explicit Type Casting:** Programmer manually converts a variable using functions like `int()`, `float()`, `str()`.


````python
# Type Checking
x = 5
y = 3.5
z = "Hello"

print(type(x))  # <class 'int'>
print(type(y))  # <class 'float'>
print(type(z))  # <class 'str'>
````

````python
# Implicit Type Casting
a = 5       # int
b = 2.5     # float

c = a + b   # int is converted to float automatically
print(c)          # 7.5
print(type(c))    # <class 'float'>
````

````python
# Explicit Type Casting
x = 10       # int

print(x, type(x))   # 10 <class 'int'>
x_float = float(x)
print(x_float, type(x_float))  # 10.0 <class 'float'>
````

## **OOP: Classes and Objects**
OOP helps us organize code using classes and objects.  
It makes code easier to understand, reuse, and manage.

### Class
A **class** is a blueprint for creating objects.  
It defines attributes (data) and methods (functions).

### Object
An **object** is an instance of a class.  
It represents a real-world entity and can access the class attributes and methods.

### Syntax:
```
class ClassName:
    def __init__(self, attribute_value):
        self.attribute = attribute_value

````python
# Basic example
class Person:
    def __init__(self, name, age):     # Constructor
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name} and I am {self.age} years old.")

# Creating an object (instance)
person1 = Person("Alice", 25)
person1.greet()
````

### **Inheritance**
Inheritance allows one class **(Child)** to use the properties and methods of another class **(Parent)**.  
It helps **reuse code** and organize related classes.

````python
class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):  # Dog inherits from Animal
    def bark(self):
        print("Dog barks")

d1 = Dog()
d1.speak()  # Inherited method from Animal
d1.bark()   # Child method
````

### **Diamond Problem**

In multiple inheritance, a class may inherit from two classes that share a common parent.  
Using `super()` ensures the common parent is called only once


````python
# Example without using super()
class Animal:
    def __init__(self):
        print("Animal init")
        self.live = True

class Aquatic(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Aquatic init")
        self.live_water = True

class Terrestrial(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Terrestrial init")
        self.live_land = True

class Amphibian(Aquatic, Terrestrial):
    def __init__(self):
        Terrestrial.__init__(self)
        Aquatic.__init__(self)
        print("Amphibian init")
        self.live_land_water = True

frog = Amphibian()
````

````python
# The same example using super()
class Animal:
    def __init__(self):
        print("Animal init")
        self.live = True

class Aquatic(Animal):
    def __init__(self):
        super().__init__()
        print("Aquatic init")
        self.live_water = True

class Terrestrial(Animal):
    def __init__(self):
        super().__init__()
        print("Terrestrial init")
        self.live_land = True

class Amphibian(Aquatic, Terrestrial):
    def __init__(self):
        super().__init__()
        print("Amphibian init")
        self.live_land_water = True


frog = Amphibian()
````

## **File Handling in Python**

File handling allows us to store and retrieve data permanently.

We will see:
- TXT file handling
- CSV file handling
- JSON file handling

We utilise the CRUD principle with file handling, ie `creating` the file and its contents, `reading` the file, `updating` content, `deleting` content.

````python
# Writing to a TXT file
with open("sample.txt", "w") as file:
    file.write("Hello Students!")

# Reading from the file
with open("sample.txt", "r") as file:
    content = file.read()

print("TXT File Content:")
print(content)

# appending
with open("sample.txt", "a") as file:
    file.write("\nThis is a new line.")

print("\nUpdated Content:")
with open("sample.txt", "r") as file:
    content = file.read()
print(content)

# changing second line

with open("sample.txt", "r") as file:
    lines = file.readlines()

lines[1] = "We Shall Watch Your Career With Great Interest"

with open("sample.txt", "w") as file:
    file.writelines(lines)

print("\nUpdated Content:")
with open("sample.txt", "r") as file:
    content = file.read()
print(content)
````

````python
import csv

students = [
    {"id": 1, "name": "Ankita", "marks": 85},
    {"id": 2, "name": "Awwab", "marks": 78}
]

# CREATE
with open("students.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["id", "name", "marks"])
    writer.writeheader()
    writer.writerows(students)

# READ
with open("students.csv", "r") as f:
    reader = csv.DictReader(f)
    data = list(reader)
    print("READ:", data)

with open("students.csv", "r") as f:
    csv_format = csv.reader(f)
    print("\nCSV Form: ")
    for row in csv_format:
      print(' , '.join(row))
# UPDATE
for row in data:
    if row["name"] == "Awwab":
        row["marks"] = "83"

# WRITE BACK
with open("students.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["id", "name", "marks"])
    writer.writeheader()
    writer.writerows(data)

# READ: AFTER UPDATE

with open("students.csv", "r") as f:
    reader = csv.DictReader(f)
    data = list(reader)
    print("\nREAD AFTER UPDATE:", data)

with open("students.csv", "r") as f:
    csv_format = csv.reader(f)
    print("\nCSV Form: ")
    for row in csv_format:
      print(' , '.join(row))
````

````python
import sqlite3

# CONNECT (creates file if not exists)
conn = sqlite3.connect("students.db")
cursor = conn.cursor()

# DROP OLD DATA IF EXISTS
cursor.execute("""
DROP TABLE IF EXISTS students
""")

# CREATE TABLE
cursor.execute("""
CREATE TABLE students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    marks INTEGER
)
""")

# INSERT
cursor.execute("INSERT INTO students (name, marks) VALUES (?, ?)", ("Ankita", 85))
cursor.execute("INSERT INTO students (name, marks) VALUES (?, ?)", ("Awwab", 78))
conn.commit()

# READ
cursor.execute("SELECT * FROM students")
print("READ:", cursor.fetchall())

# UPDATE
cursor.execute("UPDATE students SET marks = ? WHERE name = ?", (83, "Awwab"))
conn.commit()

# READ AGAIN
cursor.execute("SELECT * FROM students")
print("\nREAD AFTER UPDATE:", cursor.fetchall())

# DELETE
cursor.execute("DELETE FROM students WHERE name = ?", ("Awwab",))
conn.commit()

cursor.execute("SELECT * FROM students")
print("\nFINAL:", cursor.fetchall())

conn.close()
````

Pandas Allows for a lot of cool operations like grouping, aggregating, filtering, cleaning, etc.

It is mainly used in **Exploratory Data Analysis (EDA)** to find important features and statistical anomalies/properties of some given dataset.

````python
import pandas as pd
import numpy as np

data = {
    "Name": ["Awwab", "Avanish", "Nathan", "Mohammed"],
    "Marks": [67, 95, np.nan, -100],
    "Class": ["IT", "IT", "IT_2", "Production"],
    "Grade": ["FF","AA","AB","ZZ"]
}

df = pd.DataFrame(data,index = ["Pikachu", "Exeggutor","Grafaiai","Snorlax"]) # index is like primary key
df
````

````python
# accessing data
print(df["Marks"]) # accessing a column

print('\n',df.loc["Pikachu"]) # accessing a row via label

print('\n',df.iloc[1:4]) # accessing via integer

print('\n', df.loc[df["Name"]=="Awwab",["Name","Grade"]])
````

````python
#handling na values
print(df.isna())
print('\n')
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

print(df["Marks"])
````

````python
# updating
df.loc[df["Name"] == "Mohammed", "Marks"] = -1000 # bro is lazy
df.loc[df["Name"] == "Mohammed"]
````

````python
# aggregation
print('Mean of Marks =',df["Marks"].mean())
print('Sum of marks =',df["Marks"].sum())
# Minimum marks row
print('row with minimum marks \n ', df.loc[df["Marks"]==df["Marks"].min()])
````

````python
#grouping
df.groupby("Class")["Marks"].mean()
````

## **Matplotlib Basics**

- Matplotlib is a library used to plot stuff
- You can make like, lines , circles and even complex curves with it
- It helps in visualising the data given to us
- By making subplots you can make multiple plots under one canvas too

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Create 3 clusters
cluster1 = np.random.randn(100, 2) + [2, 2]
cluster2 = np.random.randn(100, 2) + [-2, -2]
cluster3 = np.random.randn(100, 2) + [5, -3]

plt.scatter(cluster1[:,0], cluster1[:,1],label="Cluster 1")
plt.scatter(cluster2[:,0], cluster2[:,1], label = "Cluster 2")
plt.scatter(cluster3[:,0], cluster3[:,1],label ="Cluster 3")
# random aah Line
x1 = np.linspace(-6,5,100) # start , stop, values
y1 = x1*2
plt.plot(x1,y1, label="Line aahh", color = 'purple')

plt.legend()
plt.title("Random Clusters")
plt.show()
````

````python
# some figures

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10, 4))

fig.suptitle('Line, Circle and Dil')

# line
ax1.plot(np.arange(10), np.arange(10), color='blue')
ax1.set_title('Plot 1 : y = x')
ax1.set_xlim(0,10)
ax1.set_ylim(0,20)

# circle
rad, theta = 10, np.linspace(0,2*np.pi, 100)

x2 = rad*np.cos(theta)
y2 = rad*np.sin(theta)
ax2.plot(x2,y2, color='green')
ax2.set_title('Plot 2 : x**2 + y**2 = 100')
ax2.set_xlim(-10,10)
ax2.set_ylim(-10,10)
ax2.axis('equal')

# heart
t = np.linspace(-5,5,200)

x3 = 16*(np.sin(t)**3)
y3 = 13*np.cos(t) - 5*np.cos(2*t) - 2*np.cos(3*t) - np.cos(4*t)
ax3.plot(x3, y3, color='red')
ax3.set_title('Plot 3 : Dil')
ax3.set_xlim(-20,20)
ax3.set_ylim(-20,20)
ax3.axis('equal')
plt.show()
````
